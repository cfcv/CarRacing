```python
def read_csvs(path, name):
    files = os.listdir(path)
    training_files = []
    for file in files:
        if(file.endswith("csv")):
            training_files.append(file)
    print(training_files)
    assert len(training_files) == 5, "5 csv not found"
    
    data_frames = []
    for file in training_files:
        data = pd.read_csv(path + "/" + file)
        data = data.drop(['Wall time'], axis=1)
        data["Step"] = data["Step"] / 1000000
        data = data.rename(columns={"Value": "average return", "Step": "million steps"})
        data_frames.append(data)

    data = pd.concat(data_frames, sort=False)
    data["Agent Name"] = name
    return data
```
